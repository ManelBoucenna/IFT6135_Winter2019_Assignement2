```
from matplotlib import pyplot as plt
import numpy as np
import pylab

results = np.load('learning_curves.npy')

train_ppls = results.tolist()['train_ppls']
val_ppls = results.tolist()['val_ppls']

plt.figure(figsize=(10, 6), dpi=80)
plt.plot(train_ppls, label="Train ppls", color="blue", linewidth=2.5, linestyle="-")
plt.plot(val_ppls, label="Validation ppls", color="red", linewidth=2.5, linestyle="-")
pylab.legend(loc='upper right')
plt.title('GRU_SGD_model=GRU_optimizer=SGD_initial_lr=10_batch_size=20_seq_len=35_hidden_size=1500_num_layers=2_dp_keep_prob=0.35_save_best_0')
plt.show()
```
