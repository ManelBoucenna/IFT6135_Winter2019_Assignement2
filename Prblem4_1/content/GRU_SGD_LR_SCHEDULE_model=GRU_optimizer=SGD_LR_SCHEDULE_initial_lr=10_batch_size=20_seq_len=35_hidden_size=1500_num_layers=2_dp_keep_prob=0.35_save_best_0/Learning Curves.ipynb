```python
from matplotlib import pyplot as plt
import numpy as np
import pandas
import pylab

results = np.load('learning_curves.npy')

train_ppls = results.tolist()['train_ppls']
val_ppls = results.tolist()['val_ppls']

plt.figure(figsize=(10, 6), dpi=80)
plt.plot(train_ppls, label="Train ppls", color="blue", linewidth=2.5, linestyle="-")
plt.plot(val_ppls, label="Validation ppls", color="red", linewidth=2.5, linestyle="-")
pylab.legend(loc='upper right')
plt.title('Train and Validation Perpelexities')
```

```python
train_losses = results.tolist()['train_losses']
val_losses = results.tolist()['val_losses']


plt.figure(figsize=(10, 6), dpi=80)
plt.plot(train_losses, label="Train losses", color="blue", linewidth=2.5, linestyle="-")
plt.plot(val_losses, label="Validation losses", color="red", linewidth=2.5, linestyle="-")
pylab.legend(loc='upper right')
plt.title('Train and Validation Losses')
```
