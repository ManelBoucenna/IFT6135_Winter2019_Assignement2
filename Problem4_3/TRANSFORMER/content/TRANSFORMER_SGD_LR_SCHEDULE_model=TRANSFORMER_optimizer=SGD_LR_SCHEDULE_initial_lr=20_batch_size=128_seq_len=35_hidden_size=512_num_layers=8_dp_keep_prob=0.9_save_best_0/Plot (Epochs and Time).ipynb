```python
import numpy as np
import matplotlib.pyplot as plt

plot = np.load('learning_curves.npy')
x=np.arange(0,40)

Train=plot.item()['train_ppls']
Valid=plot.item()['val_ppls']

plt.plot(x,Train,label='Train ppl')
plt.plot(x,Valid,label='Valid ppl')
plt.legend()

plt.xlabel('Epochs')
plt.ylabel('PPL')
plt.title("Transformer-SGD problem 4.3")

plt.savefig('Transformer-SGD problem 4.3(Epochs).png')
plt.show();
```

```python
# import file
import numpy as np
Train = []
Validation = []
Time = [0]

f = open("log.txt" , "r")
for line in f:
    word_list = line.split('\t')
    for word in word_list:
        line = word.split()
        if (line[0]=='train'):
            Train.append(float(line[-1]))
        else:
            if (line[0]=='val'):
                Validation.append(float(line[-1]))
            else:
                if (line[0]=='time'):
                    Time.append(Time[-1]+float(line[-1]))
                    
Train = np.asarray(Train)
Validation = np.asarray(Validation)
Time = np.asarray(Time[1:])

plt.plot(Time,Train,label='Train ppl')
plt.plot(Time,Validation,label='Valid ppl')
plt.legend()

plt.xlabel('Time')
plt.ylabel('PPL')
plt.title("Transformer-SGD problem 4.3")

plt.savefig('Transformer-SGD problem 4.3(Time).png')
plt.show();
```
